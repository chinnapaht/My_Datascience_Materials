```
# Define the headers since the data does not have any
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]

# Read in the CSV file and convert "?" to NaN
df = pd.read_csv("http://mlr.cs.umass.edu/ml/machine-learning-databases/autos/imports-85.data",
                  header=None, names=headers, na_values="?" )
df.head()
```

```
#1 Hot encoding
df_encoded=pd.get_dummies(df,prefix=["encode"],columns=["num_doors"])
df_encoded.head()
```

```
#2Label Encoding
df.loc[:,"num_door_encode"]=pd.factorize(df['num_doors'])[0].reshape(-1,1)
df
```

```
#3 Ordinal Encoding
Door_dict ={"two":2,"four":4}
df["num_doors_ordinal"]=df.num_doors.map(Door_dict)  
df
```

```
#4 Helmert Encoding 
import category_encoders as ce
encoder=ce.HelmertEncoder(cols=["num_doors"],drop_invariant=True)
dfh=encoder.fit_transform(df["num_doors"])
df=pd.concat([df,dfh],axis=1)
df
```

```
#5 Binary Encoding 
encoder=ce.BinaryEncoder(cols=["num_doors"])
dfbin=encoder.fit_transform(df["num_doors"])
df=pd.concat([df,dfbin],axis=1)
df
```

```
#6 Freq encoding
fe=df.groupby("num_doors").size()/len(df)
df.loc[:,"num_doors_freq_encode"]=df["num_doors"].map(fe)
df
```

```
#7 Mean Encoding
mean_encode=df.groupby("num_doors")["price"].mean()
print(mean_encode)
df.loc[:,"numdoors_mean_enc"]=df["num_doors"].map(mean_encode)
df
```

```
# 8 Weight Evidence Encoding
#We Calculate probability of target = 1 i.e. Good=1 for each Category
woe_df=df.groupby("num_doors")["price"].mean()
woe_df=pd.DataFrame(woe_df)
#Rename the colum to good
woe_df=woe_df.rename(columns={"price":"Good"})
#Calculate Bad Which is 1-good prob
woe_df["Bad"]=1-woe_df.Good
#We need to add a small value to avoid divide by zero denominator
woe_df["Bad"]=np.where(woe_df["Bad"] == 0 ,0.000001,woe_df["Bad"])
#compute the WOE
woe_df["WoE"] = np.log(woe_df.Good/woe_df.Bad)
woe_df
```

```
# 9 Prob Ratio Encoding
#We Calculate probability of target = 1 i.e. Good=1 for each Category
woe_df=df.groupby("num_doors")["price"].mean()
woe_df=pd.DataFrame(woe_df)
#Rename the colum to good
woe_df=woe_df.rename(columns={"price":"Good"})
#Calculate Bad Which is 1-good prob
woe_df["Bad"]=1-woe_df.Good
#We need to add a small value to avoid divide by zero denominator
woe_df["Bad"]=np.where(woe_df["Bad"] == 0 ,0.000001,woe_df["Bad"])
#compute the Probability Ratio
woe_df["WoE"] = np.log(woe_df.Good/woe_df.Bad)
woe_df
```
